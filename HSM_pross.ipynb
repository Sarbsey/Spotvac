```python
# HSM data-processing
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Flatten
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.read_csv('./data/Happy.csv')
df1 = df1.replace(to_replace="Happy", value=0, regex=True)

df2 = pd.read_csv('./data/Sad.csv')
df2 = df2.replace(to_replace="Sad", value=1, regex=True)

df3 = pd.read_csv('./data/Contentment.csv')
df3 = df3.replace(to_replace="Contentment", value=0.5, regex=True)

df = pd.concat([df1,df2])
t2 = df[['mood','danceability', 'acousticness', 'energy',
       'liveness', 'valence', 'loudness']]

t2 = t2.reset_index(drop=True)
t2.to_csv('./data/HSM_data.csv', index=False)
t2
```

```python
# load dataset
df = pd.read_csv("./data/HSM_data.csv")
df
```

```python
properties = list(df.columns.values)
properties.remove('mood')
print(properties)
X = df[properties]
y = df['mood']

print(X.head())
```

```python
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = Sequential([
    Dense(6, activation='relu', input_shape=(6,)),
	Dense(16, activation='relu'),
	Dense(16, activation='relu'),
	Dense(16, activation='relu'),
	Dense(16, activation='relu'),
	Dense(16, activation='relu'),
	Dense(16, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=200)
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test accuracy:', test_acc)
```
